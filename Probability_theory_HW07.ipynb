### Задание 1
Выбрать тест и проверить, есть  ли различия между выборками:

1 )  Даны две  независимые выборки. Не соблюдается условие нормальности

x1  380,420, 290

y1 140,360,200,900

Сделайте вывод по результатам, полученным с помощью функции

```python
from scipy import stats
import pandas as pd
import numpy as np

x1 = [380, 420, 290]

y1 = [140, 360, 200, 900]
M_W_pvalue = []
print('Используем критерий Манна-Уитни, поскольку выборки не нормальные и независимые ')
M_W_pvalue = stats.mannwhitneyu(x1, y1)
print(f'M_W_pvalue= {M_W_pvalue[1]}')
print('На уровне статистической значимости alpha = 0.05 различия м\д выборками нет ')
```

### Задание 2
Исследовалось влияние препарата на уровень давления пациентов. Сначала
измерялось давление до приема препарата, потом через 10 минут и через 30 минут. Есть
ли статистически значимые различия между измерениями давления? В выборках не соблюдается условие нормальности.

1е измерение до приема препарата: 150, 160, 165, 145, 155

2е измерение через 10 минут: 140, 155, 150, 130, 135

3е измерение через 30 минут: 130, 130, 120, 130, 125

```python
Before = [150, 160, 165, 145, 155]
Y = [140, 155, 150, 130, 135]
Z = [130, 130, 120, 130, 125]
print('Используем критерий Фридмана, поскольку имеем множество выборок для повторяемых измерений')
T, F_pvalue = stats.friedmanchisquare(Before, Y, Z)
print(f'F_pvalue= {F_pvalue}')
print('На уровне статистической значимости alpha = 0.05, нулевая гипотеза отвергается в выборках \
      имеются статистически значимые различия. Препарат влияет на давление.')
```

### Задание 3
Сравните 1 и 2 е измерения, предполагая, что 3го измерения через 30 минут не было. Есть
ли статистически значимые различия между измерениями давления?

```python
print('Используем критерий Уилкоксона, поскольку имеем 2 зависимые выборки')
T, WC_pvalue = stats.wilcoxon(Before, Y)
print(f'WC_pvalue= {WC_pvalue}')
print('На уровне статистической значимости alpha = 0.05, нулевая гипотеза не отвергается в выборках \
      нет статистически значимых различий. Препарат не влияет на давление.')
```

### Задание 4
Даны 3 группы учеников плавания. Не соблюдается условие нормальности.
В 1 группе время на дистанцию 50 м составляют: 56, 60, 62, 55, 71, 67, 59, 58, 64, 67

Вторая группа : 57, 58, 69, 48, 72, 70, 68, 71, 50, 53

Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54

Есть ли статистически значимые различия между группами?


```python
X = [56, 60, 62, 55, 71, 67, 59, 58, 64, 67]
Y = [57, 58, 69, 48, 72, 70, 68, 71, 50, 53]
Z = [57, 67, 49, 48, 47, 55, 66, 51, 54]
print('Используем критерий Крускала-Уолиса, поскольку имеем множество выборок для независимых измерений')
T, KW_pvalue = stats.kruskal(X, Y, Z)
print(f'KW_pvalue= {KW_pvalue}')
print('На уровне статистической значимости alpha = 0.05, нулевая гипотеза не отвергается в группах \
      нет статистически значимых различий.')
```

### Задание 5
Заявляется, что партия изготавливается со средним арифметическим 2,5 см. Проверить
данную гипотезу, если известно, что размеры изделий подчинены нормальному закону
распределения. Объем выборки 10, уровень статистической значимости 5%
2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34

```python
X = [2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34]
n = 10
mean = np.mean(X)
print(mean)
std = np.std(X, ddof=1)
print(std)
print('Используем критерий стьюдента поскольку среднекв отклонение ген совокупности \
неизветсно и построим доверительный интервал:')
T = stats.t.ppf(0.05/2, 10-1)
print(f'T = {T}')
X1 = mean - T*std/n**0.5
X2 = mean + T*std/n**0.5
print(f'Тогда доверительный интервал лежит в диапазоне [{X2}, {X1}]')
print('Поскольку  2,5 входит в доверительный интервал нулевая гипотеза не отклоняется, утверждение \
что партия изготавливается со средним арифметическим 2,5 см не ложно')
```
